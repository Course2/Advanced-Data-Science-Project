## Module 6: Model Development

```python
print('To avoid errors the Date format will be changed to separete Year, Month, Day')
```

```python
# To avoid errors the Date format will be changed to separete 'Year', 'Month','Day'
df1['Year'] = df['Date'].dt.year
df1['Month'] = df['Date'].dt.month
df1['Day'] = df['Date'].dt.day
```

```python
#final dataset
dataset = df1[['Day', 'Month', 'Year', 'High', 'Open', 'Low', 'Close']]
dataset.head()
```

```python
dataset.tail()
```

##### Spliting the data into training set and test set

```python
# Separate data as independent and dependent variables
X = dataset.iloc[:,dataset.columns !='Close']
y = dataset.iloc[:, 5]

print(X.shape)
print(y.shape)
```

```python
# Split dataset into train and test sets
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=.25)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Module 7 & 8: Building and Evaluating Models

```python
print("Here we built and evaluate the Linear Models.  They are LinearRegression model, Support Vector Machine Model, K-Nearest Neighbor Model.  Cross validation is used to measure accuracy on the Linear Regression Model and the KNN Model.")
```

###### Model-1 Linear Regression

```python
# Linear Regression Model Training and Testing
from sklearn.linear_model import LinearRegression
```

```python
# Accuracy_score
from sklearn.metrics import accuracy_score
```

```python
# Linear Regression Model Training and Testing

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_predict=lr_model.predict(X_test)
```

```python
# Plotting Actual vs predicted
plot_dataset = pd.DataFrame({'Actual':y_test,'Pred':y_predict})
plot_dataset.tail(50).plot(kind='bar',figsize=(15, 6))
plt.title("Linear Regression Model ")
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import math
```

```python
lr_mse=math.sqrt(mean_squared_error(y_test,y_predict))
print('Linear Model root mean square error', lr_mse)
```

```python
LR_R2=r2_score(y_test, y_predict)
print('Linear Model R2 score', LR_R2)
```

```python
lr_mae = mean_absolute_error(y_test, y_predict)
print('Linear Model Mean Absolute Error', lr_mae)
```

#### Linear Cross-Validation

```python
print("Cross-Validation is a technique where the model is evaluated on a dataset on which it is not trained.  It can be the test data or another dataset.")
```

```python
from sklearn import model_selection 
from sklearn.model_selection import KFold
```

```python
results = model_selection.KFold(n_splits=25)
results
```

```python
results = model_selection.cross_val_score(lr_model, X_test, y_test.astype('int'), cv=results)
```

```python
print("Accuracy: ", results.mean()*100)
```

#### Model-2 Support Vector Machine(SVM)

```python
from sklearn.svm import SVR
```

```python
svm_regressor = SVR(kernel='linear')
svm_model=svm_regressor.fit(X_train, y_train)
y_svm_predict=svm_model.predict(X_test)
```

```python
# Plot svm actual vs predict
plot_svm_dataset = pd.DataFrame({'Actual':y_test,'Pred':y_svm_predict})
plot_svm_dataset.tail(50).plot(kind='bar',figsize=(15, 6))
plt.title("SVM Model ")
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
svm_mse=math.sqrt(mean_squared_error(y_test,y_svm_predict))
print('SVM Model root mean square error', svm_mse)
```

```python
SVM_R2=r2_score(y_test, y_svm_predict)
print('SVM Model R2 score', SVM_R2)
```

```python
SVM_mae = mean_absolute_error(y_test, y_svm_predict)
print('SVM Model Mean Absolute Error', SVM_mae)
```

#### Model-3 K-Nearest Neighbour(KNN)

```python
from sklearn.neighbors import KNeighborsRegressor
knn_regressor=KNeighborsRegressor(n_neighbors=4)
knn_model=knn_regressor.fit(X_train, y_train)
y_knn_predict=knn_model.predict(X_test)
```

```python
# Plot for actual vs predict in KNN
plot_knn_dataset = pd.DataFrame({'Actual':y_test,'Pred':y_knn_predict})
plot_knn_dataset.tail(50).plot(kind='bar',figsize=(15, 6))
plt.title("K-Nearest Neighbour Model ")
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
KNN_mse=math.sqrt(mean_squared_error(y_test,y_knn_predict))
print('KNN Model root mean square error', KNN_mse)
```

```python
KNN_R2=r2_score(y_test, y_knn_predict)
print('KNN Model R2 score', KNN_R2)
```

```python
KNN_mae = mean_absolute_error(y_test, y_knn_predict)
print('KNN Model Mean Absolute Error', KNN_mae)
```

#### KNN Cross-Validation

```python
knn_kfold = model_selection.KFold(n_splits=25)
knn_kfold
```

```python
results_kfold = model_selection.cross_val_score(knn_model, X_test, y_test.astype('int'), cv=knn_kfold)
```

```python
print("Accuracy: ", results_kfold.mean()*100)
```
